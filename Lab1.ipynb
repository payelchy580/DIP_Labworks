```python
import os
imagefolder = "/content/drive/MyDrive/Colab Notebooks/DIP Lab Works/rabbit"
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# List all files in the folder
images = os.listdir(imagefolder)

# Display one image
img_path = os.path.join(imagefolder, images[200])
img = mpimg.imread(img_path)

plt.imshow(img)
plt.axis('off')
plt.title('Sample Image from Google Drive')
plt.show()
# Get dimensions
height, width, channels = img.shape

# Get image size in bytes
image_size = img.size  # total number of pixels Ã— channels
image_dtype = img.dtype  # data type (usually uint8)

print("Image Details")
print(f"Width: {width} pixels")
print(f"Height: {height} pixels")
print(f"Image Resolution (in pixels): {image_size}")
print(f"Data Type: {image_dtype}")
```
